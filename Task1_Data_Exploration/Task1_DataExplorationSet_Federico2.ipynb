# Statistics per Category

```python
rtbCost = rtb.drop(columns=['bidprice','hour','bidid','userid','useragent','IP','region','city','adexchange','domain','url','urlid','slotid','slotwidth','slotheight','slotvisibility','slotformat','slotprice','creative','keypage','advertiser','usertag'])

# Average cost for impressions for a given WEEKDAY
rtbCostC = rtbCost.loc[rtbCost['payprice']]
rtbCostFinal = rtbCostC.groupby(['weekday']).mean()
rtbCostFinal
```

```python
#Table ordered by Clicks per Weekday
rtbWe = rtb.drop(columns=['bidprice','hour','bidid','userid','useragent','IP','region','city','adexchange','domain','url','urlid','slotid','slotwidth','slotheight','slotvisibility','slotformat','slotprice','creative','keypage','advertiser','usertag'])
rtbWeClick = rtbWe.loc[rtbWe['click'] ==1 ]
rtbWeC = rtbWeClick.groupby(['weekday']).size().reset_index(name = 'clicks')

#table ordered by payprice per Weekday
#(['bidprice','weekday','hour','bidid','userid','useragent','IP','region','city','adexchange','domain','url','urlid','slotid','slotwidth','slotheight','slotvisibility','slotformat','slotprice','creative','keypage','advertiser','usertag'])

#Impressions per weekday
impsWe = rtbWe.groupby(['weekday']).size().reset_index(name = 'impressions')
#CTR per weekday
CTR =  rtbWeC ['clicks'] / impsWe['impressions']
impsWe ['CTR'] = CTR.values 
#clicks per weekday
clicks = CTR * impsWe['impressions']
impsWe ['clicks'] = clicks.values
#average pay price per weekday

impsWe ['avgImpCost'] = ""
impsWe ['avgImpCost'] = rtbCostFinal ['payprice']
impsWe

```

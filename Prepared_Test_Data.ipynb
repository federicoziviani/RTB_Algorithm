```python
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Imputer
%matplotlib inline
```

```python
#Load data
df_test = pd.read_csv("test.csv")
```

```python
df_test_bidid = df_test['bidid']
df_test_bidid.to_pickle('df_test_bidid')
df_test = df_test.drop(['urlid','bidid', 'userid'], axis = 1)
```

```python
df_test_bidid
```

```python
# Check for null in df 

df_test_cols = list(df_test)
for col_name in df_test_cols:
    print(col_name, df_test[col_name].isnull().values.any())
```

```python
# Fill null as 'unknown' 
df_test['url'].fillna('unknown', inplace = True)
df_test['keypage'].fillna('unknown', inplace = True)
df_test['usertag'].fillna('unknown', inplace = True)

# Convert adexchange data type to int 
df_test['adexchange'].fillna(0 , inplace = True )
df_test.adexchange =  df_test.adexchange.astype(int)

# Convert domain data type 
df_test.domain = df_test.domain.astype(str)

# Check for null in df again
col_names = list(df_test)
for column in col_names:
    print(column, df_test[column].isnull().values.any())
```

```python
lb_encode = LabelEncoder()

test_cat_cols = ['useragent','IP','url','domain','slotid','slotvisibility','slotformat','creative','keypage','usertag']
for cat_name in test_cat_cols:
    print('endcoding', cat_name, '.....')
    df_test[cat_name + '_encoded'] = lb_encode.fit_transform(df_test[cat_name])

print()
print('Encoding done!')
```

```python
df_test_prepared = df_test.drop(columns = ['useragent','IP','domain','url','slotid','slotvisibility','slotformat','creative','keypage','usertag'])

# Save prepared df 
df_test_prepared.to_pickle('df_test_prepared')

df_test_prepared.head()
```

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
pCTR_train = pd.read_pickle('pCTR_validation')
rtb_validation = pd.read_csv('validation.csv')
```

# ORTB using Ada Boost for CTR predicion

```python
ada_boost_val = pd.read_pickle('ada_boost_val')
print(ada_boost_val.head())
bidPrices = []
lagr = 0.001
c = 70
```

```python
for i in range(0, len(rtb_validation)):
    price = np.sqrt((c/lagr)*ada_boost_val['pCTR'].values[i] + c**2) - c
    bidPrices.append(price)
```

```python
rtb_validation.head()
```

```python
#run the algorithm to sepend the budget

budget = 6250
impressions = 0
bids = 0
clicks = 0
for i in range(0, len(rtb_validation)):
    if budget > 0 and budget > (rtb_validation['payprice'].values[i]/1000):
        impressions += 1
        if (bidPrices[i]) > (rtb_validation['payprice'].values[i]):
            budget = budget - (rtb_validation['payprice'].values[i]/1000)
            bids += 1
            clicks += rtb_validation['click'].values[i] 
    else:
        print("Budget is spent!")
        break
consumedBudget = 6250 - budget
ctr = (clicks/bids)*100
cpm = ((consumedBudget)/bids)*1000
cpc = (consumedBudget)/clicks 
#Output for the alrogithm
print("Output")
print("Number of Bids:",bids)
print("Number of Impressions:",impressions)
print("Number of Clicks:", clicks)
print("CTR is:",ctr, "%")
print("CPM is:",cpm)
print("CPC is:",cpc)
print (budget)
print(consumedBudget)
```
